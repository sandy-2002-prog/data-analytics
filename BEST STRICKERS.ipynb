# Statistical analysis

### Find whelther there is significant difference in consistency rates among strikers from different nationality

```python
# Normality test
from scipy.stats import shapiro
stat,p_value=shapiro(data['Consistency'])
print("P value: ",round(p_value,3))
```

```python
#Filtering test
Spain=data.query('Nationality == "Spain"')['Consistency']
France=data.query('Nationality == "France"')['Consistency']
Germany=data.query('Nationality == "Germany"')['Consistency']
Brazil=data.query('Nationality == "Brazil"')['Consistency']
England=data.query('Nationality == "England"')['Consistency']
```

```python
#Levene test
from scipy.stats import levene
stats,p_value=levene(Spain,France,Germany,Brazil,England)
print("P value: ",round(p_value,3))
```

```python
# One way ANOVA
from scipy.stats import f_oneway
Test_stat,p_value=f_oneway(Spain,France,Germany,Brazil,England)
print("P value: ",round(p_value,2))
```

# Feature Engineering

```python
data['Total Contribution Score']=(data['Goals Scored']+data['Assists']+data['Shots on Target']+data['Dribbling Success']
                                  +data['Aerial Duels Won']+data['Defensive Contribution']+data['Big Game Performance']
                                  +data['Consistency'])
```

### Encode the Footedness and Marital status by LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
data['Footedness']=encoder.fit_transform(data['Footedness'])
data['Marital Status']=encoder.fit_transform(data['Marital Status'])
data.head()
```

### Create dummies for nationality and add with data


```python
dummies=pd.get_dummies(data['Nationality'])
processsed_df=pd.concat([data,dummies],axis=1)
processed_df=processsed_df.drop('Nationality',axis=1)
processed_df.head()
```

# Cluster Analysis

### Perform K-Means clustering

```python
#Select features
x=processed_df.drop('Striker_ID',axis=1)
```

```python
import os
os.environ["OMP_NUM_THREADS"]='1'
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,15):
    kmeans=KMeans(n_clusters=i,init='k-means++')
    kmeans.fit(x)
    wcss_score=kmeans.inertia_
    wcss.append(wcss_score)
```

```python
#elbow chart
plt.figure(figsize=(12,6))
plt.plot(range(1,15),wcss,marker='o')
plt.title('Elbow Methods')
plt.xlabel('No of clusters(k)')
plt.ylabel('wcss')
plt.show()
```

```python
#Build k-means with k=2
final_km=KMeans(n_clusters=2)
final_km.fit(x)
labels=final_km.labels_
labels
```

```python
#Add lables
processed_df['Clusters']=labels
processed_df.head()
```

```python
#Checking clusters
round(processed_df.groupby('Clusters')['Total Contribution Score'].mean(),2)
```

```python
#Assign names
mapping={0:'Best Strikers',1:'Regular Strikers'}
processed_df['Strikers_types']=processed_df['Clusters'].map(mapping)
```

```python
#Delate clusters variables
processed_df=processed_df.drop('Clusters',axis=1)
processed_df.head()
```

# Data preprocessing for ML

### New feature mapping

```python
mapping={'Best Strikers':1,'Regular Strikers':0}
processed_df['Strikers_types']=processed_df['Strikers_types'].map(mapping)
processed_df.head()
```

### Selecting Features

```python
x=processed_df.drop(['Striker_ID','Strikers_types'],axis=1)
y=processed_df['Strikers_types']
```

### Scaling features

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_x=scaler.fit_transform(x)
scaled_x
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(scaled_x,y,test_size=0.2,random_state=42)
```

# Predictive Classification analysis

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix
#Model_train
lgr_model=LogisticRegression()
lgr_model.fit(x_train,y_train)
#Prediction
y_lgr_pred=lgr_model.predict(x_test)
#Evaluation
accuracy_lgr=accuracy_score(y_test,y_lgr_pred)
print(accuracy_lgr*100,'%')
```

```python
#confusion matrix
confusion_matrix_lgr=confusion_matrix(y_test,y_lgr_pred)
```

```python
#Plot confusion matrix
plt.figure(figsize=(12,6))
sns.heatmap(confusion_matrix_lgr,annot=True,fmt="d",cmap="Blues")
plt.title('Confusion Matrix for lgr model')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
